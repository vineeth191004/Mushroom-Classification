```python
data.isnull().sum() # calculates number of null values for each column
```

```python
from sklearn.preprocessing import LabelEncoder

def label_encoded(feat):
  le = LabelEncoder()
  le.fit(feat)
  print(feat.name,le.classes_)
  return le.transform(feat)
```

```python
for col in data.columns:
  data[str(col)] = label_encoded(data[str(col)])
```

```python
# cap-shape ['b' 'c' 'f' 'k' 's' 'x'] -> [0, 1, ,2 ,3 ,4, 5] -> label encoder working principle
```

```python
y = data['class']
X = data.drop(['class'], axis = 1)
```

```python
# So we see a correlation value in the range of (-0.53 to 0.9) exculding the 1s. For removing some of the columns
# so that our dataset is smaller in dimesion and yet still holds most of the information, we can remove the highly
# correlated columns. A better way is to use PCA, which calculates linear combinations of your variables in such way
# that they get "summarized" in fewer variables.

from sklearn.decomposition import PCA

pca = PCA(n_components = 6)
pca2 = PCA(n_components = 16)
pca_fit = pca.fit_transform(X)
pca_fit2 = pca2.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(pca_fit, y, test_size=0.2, random_state = 40)
```

```python
random_model = RandomForestClassifier(n_estimators=200, n_jobs = 1)
```

```python
#Fit
random_model.fit(Xtrain, ytrain)

y_pred = random_model.predict(Xtest)

#Checking the accuracy
random_model_accuracy = round(random_model.score(Xtrain, ytrain)*100,2)
print(round(random_model_accuracy, 2), '%')
```

```python
random_model_accuracy1 = round(random_model.score(Xtest, ytest)*100,2)
print(round(random_model_accuracy1, 2), '%')
```
